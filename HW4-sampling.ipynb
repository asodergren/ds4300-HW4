```python
import pandas as pd
```

```python
df_spotify = pd.read_csv("spotify.csv")
df_spotify.head()
```

```python
df_spektor = df_spotify.loc[df_spotify['artists'] == "Regina Spektor"]
df_spektor.head()
```

```python
df_spotify.drop(df_spektor.index, inplace=True)
```

```python
df_spotify_sampled = df_spotify.sample(n=1197)
```

```python
df_sampled = pd.concat([df_spotify_sampled, df_spektor], ignore_index = True)
```

```python
df_sampled.to_csv("spotify-sampled")
```
